```python
def best_model(a,b):
    algos={
        'linear regression':{
            'model':LinearRegression(),
            'param':{
                'normalize':[True,False]
            }
        },
        'lasso':{
            'model':Lasso(),
            'param':{
                'alpha':[1,2],
                'selection':['random','cyclic']
            }
        },
        'DecisionTreeRegressor':{
            'model':DecisionTreeRegressor(),
            'param':{
                'criterion':['mse','friedman_mse'],
                'splitter':['best','random']
            }
        },
        'XGBRegressor':{
            'model':XGBRegressor(),
            'param':{}
        }
    }
    scores=[]
    ss=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
    for algos_name,para in algos.items():
        gs=GridSearchCV(para['model'],para['param'],cv=ss,return_train_score=False)
        gs.fit(a,b)
        scores.append({
            'Model':algos_name,
            'Best Score':gs.best_score_,
            'Best Params':gs.best_params_
        })
    return pd.DataFrame(scores,columns=['Model','Best Score','Best Params'])
```

```python
best_model(X,Y)
```

```python
def prediction():
    print('------------AREA-----------\n1. Adyar\n2. Anna Nagar\n3. Chrompet\n4. KK Nagar\n5. Karapakkam\n6. T Nagar\n7. Velachery')
    
    a={'Adyar':0,'Anna Nagar':1,'Chrompet':2,'KK Nagar':3,'Karapakkam':4,'T Nagar':5,'Velachery':6}
    areas=input('Enter the Area : ')
    
    if areas not in a:
        print('Please enter valid Area')
    else:    
        
        area=a[areas]
        sqft=int(input('Enter Total SQ.FEET : '))
        n_bed=int(input('Number of Bedrooms : '))
        n_bath=int(input('Number of Bathrooms : '))
        n_room=int(input('Number of Rooms : '))
        
        p={'yes':1,'no':0}
        park=(input('You need a Parking ? (yes/no) :'))
        
        if park not in p:
            print('Please enter valid Parking option')
        else:    
            
            parking=p[park]
            bt={'commercial':0,'house':1,'others':2}
            typ=input('Type needed (commercial/house/others) : ')
            
            if areas not in a:
                    print('Please enter valid build type')
            else:        
                btype=bt[typ]
                reg_fee=int(input('Enter the Registeration Fee : '))
                commis=int(input('Enter the Commission : '))
                age=int(input('Enter the age of the building : '))
                
                result=dtr.predict([[area,sqft,n_bed,n_bath,n_room,parking,btype,reg_fee,commis,age]])[0]
                
                print("Your Estimation for your house :",(result)/100000,'lakhs')
```
